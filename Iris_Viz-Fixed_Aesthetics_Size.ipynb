```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
iris=pd.read_csv("iris.csv")
```

```python
iris.info()
```

```python
iris.head()
```

```python
iris.drop('Id',axis=1,inplace=True)
```

```python
iris.Species.value_counts()
```

```python
# Aesthetics !!!
sns.set_style('whitegrid')
sns.set_palette('Accent')
```

```python
#print(plt.style.available)
#plt.style.use('seaborn-muted')
```

```python
# Intial summary of data
iris.describe().plot(kind='area')
iris.describe()
```

```python
# Count Plot for species # Countplot
sns.countplot('Species',data=iris)
```

```python
# Pie chart for species
iris['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(7,5))
```

```python
# Joint Plots for 2 variables and the study of their relationship
# Different Kinds for this relation plots
```

```python
sns.jointplot('SepalLengthCm','SepalWidthCm',data=iris)
```

```python
sns.jointplot('SepalLengthCm','SepalWidthCm',data=iris,kind='reg')
```

```python
sns.jointplot('SepalLengthCm','SepalWidthCm',data=iris,kind='hex')
```

```python
sns.jointplot('SepalLengthCm','SepalWidthCm',data=iris,kind='kde',space=0,color='g')
```

```python
sns.jointplot('SepalLengthCm','SepalWidthCm',data=iris,color='k').plot_joint(sns.kdeplot,zorder=0,n_levels=6)
```

```python
# Box and violen plots
sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
```

```python
sns.violinplot(x='Species',y='SepalWidthCm',data=iris,inner='points')
```

```python
# for all numeric variables by species
iris.boxplot(by='Species',figsize=(12,6))
```

```python
# All pairplots and the distributions of single variables 
sns.pairplot(iris,hue='Species')
```

```python
#Heat map: Find the correlation
iris.corr()
sns.heatmap(iris.corr(),annot=True)
```

```python
# Andrews Curve scatter matrix, radviz, parallel co-ords
from pandas.plotting import andrews_curves,scatter_matrix,radviz,parallel_coordinates
```

```python
# Parallel Co-ordinate :for plotting multivariate, numerical data. 
                        #Parallel Coordinates Plots are ideal for comparing many variables together & 
                        #seeing the relationships between them.
        
parallel_coordinates(iris,'Species',colormap='cool')
```

```python
andrews_curves(iris,'Species',colormap='Accent')
```

```python
# Radviz: multivariate data visualization algorithm that plots each feature dimension uniformly around the 
        # circumference of a circle then plots points on the interior of the circle such that the point normalizes
        #its values on the axes from the center to each arc.
        
radviz(iris, "Species")
```
